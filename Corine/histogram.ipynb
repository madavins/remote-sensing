```python
import numpy as np
import matplotlib.pyplot as plt
import os
import numpy as np
import pandas as pd
```

```python
count = np.zeros(14)
root = '...'

i = 0
for image in os.listdir(root): 
    i = i+1
    x = root + '/' + image
    if x.endswith('.npy'):
        label = np.load(x)

        hist, bins = np.histogram(label.ravel(),14,[0,13], density=True)
        count += hist
```

```python
classes = ["Clouds","Artificial surfaces and constructions","Cultivated areas",
                       "Vineyards","Broadleaf tree cover",
                      "Coniferous tree cover","Herbaceous vegetation",
                      "Moors and Heathland","Sclerophyllous vegetation",
                     "Marshes","Peatbogs","Natural material surfaces",
                      "Permanent snow covered surfaces",
                      "Water bodies"
                       ]

plt.figure(figsize = (8,5))
plt.title("Train set class distribution")
plt.xticks(rotation=80)
plt.bar(classes, count);
```

```python
pd.set_option('display.float_format', lambda x: '%.0f' % x)

df = pd.DataFrame(count, classes)
df.columns =['Total pixels']
```
