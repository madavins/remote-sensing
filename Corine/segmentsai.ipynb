```python
dataset = "madavins/train_final"
release = client.get_release(dataset, 'v0.1')
dataset = SegmentsDataset(release, filter_by='unlabeled')

# Visualize a few samples in the dataset
samples = []
for sample in dataset:
    print(sample['name'])   
    samples.append(sample)

for sample in dataset:
    #Using the name of the image from the dataset we fetch the numpy matrix.
    file_name_npy = sample['name'].replace('.png', '.npy')
    segmentation_bitmap = np.load('./data/train_labels/' + file_name_npy)
    
    # Convert the segmentation map to the required format.
    file = bitmap2file(np.uint32(segmentation_bitmap)) # Note that this needs a np.uint32 array as input.
    asset = client.upload_asset(file, "label.png") # Name doesn't matter here.
    
    # Now add the label to a sample through the api.
    sample_uuid = sample['uuid']
    labelset = "ground-truth"

    #Required format to upload the label
    annotations = []
    for value in np.unique(segmentation_bitmap):
        annotations.append({
            "id": int(value),
            "category_id": int(value)
        })
    
    attributes = {
        "format_version": "0.1",
        "annotations": annotations,
        "segmentation_bitmap": {
            "url": asset["url"] # This refers to the previous cell
        },
    }
    #Upload the label (attributes) linking it with the image from the dataset (sample_uuid)
    client.add_label(sample_uuid, labelset, attributes)
```
